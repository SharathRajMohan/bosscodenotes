### In Class Notes

- Basic Terminologies:
    - Statistics: mathematical characteristics of samples
    - Parameters: mathematical characteristics of the population
    - Statistics is used to estimate parameters
- Inferential statistics involves making predictions or inferences about a population based on a sample of data drawn from it. Unlike descriptive statistics, which merely describes the data at hand, inferential statistics allows you to draw conclusions and make decisions.
    - Population: A population is the entire group that you want to draw conclusions about.
    - Sample: A sample is the specific group that you will collect data from. The size of the sample is always less than the total size of the population.
        - Must be representative of the population: The sample has the same characteristics as the population.
        - If every case in the population has the same chance of being selected, the sample is likely to be representative.
        - EPSEM: Equal probability of selection method.
- Inferential statistics can be considered as deriving reliable estimations from studies made on the sample of the population.
- Estimator: A formula or rule used to calculate an estimate.
- Estimate: The specific value obtained using the estimator. 
- Estimations in Inferential Statistics:
      - Estimation is a fundamental technique in inferential statistics that involves using sample data to make educated guesses about population parameters. It's a way to bridge the gap between what we know about a sample and what we want to learn about the population from which it came.
      - Point estimation and interval estimation, and hypothesis testing are three main ways of learning about the population parameter from the sample statistic.
      - An estimator is particular example of a statistic, which becomes an estimate when the formula is replaced with actual observed sample values.
         1. Point Estimation: a single value that estimates the parameter. Point estimates are single values calculated from the sample
         2. Range Estimation: Range estimation is a method used in inferential statistics to provide a range of values within which a population parameter is likely to lie. This concept is often synonymous with interval estimation, where the goal is to establish a range (interval) that is believed to contain the true value of the parameter with a certain level of confidence.
         3. Interval Estimation
               - An interval estimate provides a range of values within which the population parameter is likely to fall.
               - A common example is a confidence interval, which gives a range of values with a certain level of confidence. 
               - Confidence Interval (CI): The range of values within which the parameter is expected to lie.
               - Confidence Level: The probability that the confidence interval contains the true population parameter, typically expressed as a percentage (e.g., 95%, 99%).
               - For instance, a 95% confidence interval means there's a 95% probability that the true population parameter lies within that interval.
- Margin of Error(alpha): The maximum expected difference between the point estimate and the actual population parameter.
- Hypothesis tests: tests for a specific value(s) of the parameter.
- In order to perform these inferential tasks, i.e., make inference about the unknown population parameter from the sample statistic, we need to know the likely values of the sample statistic. What would happen if we do sampling many times?
- Sampling:
   - Sampling is a fundamental process in inferential statistics, where a subset of a population is selected to represent the entire group. 
   - The sample is the foundation of estimation. The quality and representativeness of the sample directly influence the accuracy and reliability of the estimates. The choice of sampling technique significantly impacts the accuracy and reliability of the inferences drawn.

      - Probability Sampling AKA Random Sampling:

      - Non probability Sampling
- Sampling Distribution: The sampling distribution is a probability distribution of a statistic, considering that a statistic is a random variable, or a function of a sample.
-  The standard error of a statistic is the standard deviation of its sampling distribution - it measures the variability or dispersion of the statistic due to sampling from a population.
