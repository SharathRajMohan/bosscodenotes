- Basic concepts of probability:
    - Experiment: Action with an outcome. Eg. Tossing a coin.
    - Random Experiment: An experiment with an uncertainty of the outcome. The outcome is random.
    - Event: Specific outcome or group of outcomes from the sample space of outcomes of the experiment.
        - Simple Events: Single outcome.
        - Compound outcomes: When there are multiple events for the experiment.
        - Certain event: When the probability is 1.
        - Impossible Event: When the probability is 0.
        - Mutually exclusive Event: Events that can not occur at the same time.
        - Exclusive events: Events that can occur together at the same time.
        - Sample Space: All possible outcomes of an experiment

    - Independent Events: The outcome of one event is not effecting the outcome of another event.
    - Dependent Events: The outcome of one event is effecting the outcome of subsequent event.
    - Law of total probability: 
        - Suppose ùêµ1,B2..Bn are mutually exclusive and exhaustive events (i.e., they don't overlap and they cover the entire sample space), and let A be any event. Then:
        - 
        - You can find the probability of an event A by:

            1. Breaking down the situation into several distinct cases (e.g., whether a person is sick or not).

            2. For each case, calculate:
                - The probability of that case (e.g., probability person is sick)

                - The probability of A happening given that case (e.g., probability of a positive test given the person is sick)

            3. Multiply those two values for each case and add them up
    - Bayes Theorem: 
    - Random Variable: A variable which will represent the outcome of a random experiment.
        - Discrete Random Variable: Fixed set of outcomes.
        - Continuous Random Variable: Outcomes that can span over a range of values.
    - Expectation (Mean): Expectation is a measure of the central tendency of a random variable. It represents the average value that a random variable is expected to take over many trials.
    - Variance: Variance is a measure of how spread out the values of a random variable are. It quantifies the dispersion of the data around the mean.
        - How far the values are distributed from the expected value.

    - Covariance: Covariance is a measure of how much two random variables change together. It indicates the direction of the linear relationship between the variables: whether they tend to increase or decrease together or whether one variable tends to increase when the other decreases.
    - Properties of Covariance
        - Positive Covariance: Indicates that as ùëã X increases, ùëå Y tends to increase as well.
        - Negative Covariance: Indicates that as ùëã X increases, ùëå Y tends to decrease.
        - Zero Covariance: Indicates that there is no linear relationship between ùëã X and ùëå Y.

## üîπ Bayes‚Äô Theorem

**Formula**:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

**Use Case**: Update the probability of an event (e.g. disease) given new evidence (e.g. test result).

**Law of Total Probability**:

$$
P(B) = P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A)
$$

Used to calculate $P(B)$ when there are multiple ways $B$ can occur.

---

## üîπ Z-score (Standardization)

**Formula**:

$$
z = \frac{x - \mu}{\sigma}
$$

* Tells you how many **standard deviations** a value is from the **mean**.
* Converts any normal distribution to the **standard normal distribution** $\mathcal{N}(0, 1)$.

**Interpretation**:

* $z = 0$: exactly at the mean
* $z = 1$: 1 SD above the mean
* $z = -2$: 2 SDs below the mean

---

## üîπ Standard Normal Table (Z-table)

* Provides: $P(Z \leq z)$, the **cumulative probability** up to a z-score.
* To find $P(Z > z)$:

  $$
  P(Z > z) = 1 - P(Z \leq z)
  $$

**Example**:
If $z = 0.83$:
From the table, $P(Z \leq 0.83) \approx 0.7967$
Then, $P(Z > 0.83) = 1 - 0.7967 = 0.2033$

---

## üîπ Sum of Independent Normal Variables

If $X \sim \mathcal{N}(\mu_X, \sigma_X^2)$,
and $Y \sim \mathcal{N}(\mu_Y, \sigma_Y^2)$,
and $X, Y$ are **independent**, then:

$$
X + Y \sim \mathcal{N}(\mu_X + \mu_Y, \sigma_X^2 + \sigma_Y^2)
$$

---

## üîπ Steps to Solve Normal Distribution Problems

1. **Find distribution** of the variable (mean, variance)
2. **Convert to z-score**:

   $$
   z = \frac{x - \mu}{\sigma}
   $$
3. **Look up** $z$ in the **standard normal table**
4. **Interpret** the result depending on the inequality (>, <, etc.)
